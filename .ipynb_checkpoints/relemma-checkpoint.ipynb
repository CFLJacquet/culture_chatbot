```python
import pandas as pd

a = pd.read_csv("lemmata_limsi.txt", sep = "	")

a[a.duplicated(subset = "$", keep=False)].head(10)
```

```python
b = a.drop_duplicates(subset="$", keep='last', inplace=False)
```

```python
import numpy as np
b['$'] = b['$'].str.lower()
b['='] = np.where(b['='] == "=", b['$'], b['='])
b['let1'] = b['$'].str[0]
b
```

```python
b['let1'].unique()
```

```python
pos = pd.read_csv("pos.csv", sep=';')
pos.head()
```

```python
m = b.join(pos.set_index('code'), on ='Xu')
m.to_csv("lemmata_limsi_clean.csv")
```

```python
m['info'] = m.apply(lambda row: (row['$'], row['='], row['category']), axis=1)
m.drop(['$', 'Xu', '=', 'category'], axis=1, inplace = True)
```

```python
for letter in m["let1"].unique():
    if letter not in ['£', 'é', 'à', 'â', 'ç', 'è', 'ê', 'î', 'ï', 'ô']:
        pass
        #result = m[m["let1"] == letter].drop('let1', axis=1)
        #result.to_json("lemma/first_letter_{}.json".format(letter), orient='values')
    elif letter == '£':
        result = m[m["let1"] == letter].drop('let1', axis=1)
        result.to_json("lemma/first_letter_pound.json", orient='values')        
    elif letter == 'é':
        result = m[m["let1"] == letter].drop('let1', axis=1)
        result.to_json("lemma/first_letter_e_up.json", orient='values')
    elif letter == 'à':
        result = m[m["let1"] == letter].drop('let1', axis=1)
        result.to_json("lemma/first_letter_a_down.json", orient='values')
    elif letter == 'â':
        result = m[m["let1"] == letter].drop('let1', axis=1)
        result.to_json("lemma/first_letter_a_circ.json", orient='values')
    elif letter == 'ç':
        result = m[m["let1"] == letter].drop('let1', axis=1)
        result.to_json("lemma/first_letter_c_ced.json", orient='values')
    elif letter == 'è':
        result = m[m["let1"] == letter].drop('let1', axis=1)
        result.to_json("lemma/first_letter_e_down.json", orient='values')
    elif letter == 'î':
        result = m[m["let1"] == letter].drop('let1', axis=1)
        result.to_json("lemma/first_letter_i_circ.json", orient='values')
    elif letter == 'ï':
        result = m[m["let1"] == letter].drop('let1', axis=1)
        result.to_json("lemma/first_letter_i_trema.json", orient='values')
    elif letter == 'ô':
        result = m[m["let1"] == letter].drop('let1', axis=1)
        result.to_json("lemma/first_letter_o_circ.json", orient='values')
        
```
